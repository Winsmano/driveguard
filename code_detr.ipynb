```python
from ultralytics import RTDETR, YOLO
```

```python
model_detect = RTDETR('best (4)_detect.pt')
# model_classify = YOLO('best (4)_classify.pt')
model_classify = YOLO('drive_guard_best.pt')
```

```python
import cv2
import numpy as np

def process_video():
    cap = cv2.VideoCapture(0)
    
    if not cap.isOpened():
        print("Error: Could not open camera")
        return

    try:
        while True:
            ret, frame = cap.read()
            
            if not ret:
                print("Error: Can't receive frame")
                break
            
            # Resize input frame to 640x640
            # Convert to grayscale
            gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            gray_frame = cv2.resize(frame, (1024, 1024))
            
            try:
                # Get model predictions
                results = model_classify(frame)
                
                # Plot results on the frame
                annotated_frame = results[0].plot()
                
                # Display the annotated frame
                cv2.imshow('frame', annotated_frame)
                
            except Exception as e:
                print(f"Error processing frame: {str(e)}")
                cv2.imshow('frame', gray_frame)

            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
                
    finally:
        cap.release()
        cv2.destroyAllWindows()

if __name__ == "__main__":
    process_video()

```
